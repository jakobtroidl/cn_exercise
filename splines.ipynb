# Cubic Spline Interpolation

```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt
```

```python
def myfunc(x): return 1/(1+x*x)

pts = np.linspace(-5, 5, 21)
vals = myfunc(pts)
print (pts, vals)
```

```python
tck = interpolate.splrep(pts, vals)
x = np.linspace(-5,5,100)
plt.plot (x, myfunc(x), x, interpolate.splev(x, tck), pts, vals, "x")
```

compare spline interplant with polynomial interpolation

```python
poly = interpolate.lagrange(pts, vals)
plt.ylim(-1, 2)
plt.plot (x, myfunc(x), x, interpolate.splev(x, tck), x, poly(x), pts, vals, "x")
```

spline interpolation easily allows local refinement of knots:

```python
a = 100
pts = a * np.linspace(-1,1,30)**3
vals = myfunc(pts)
tck = interpolate.splrep(pts, vals)

x = np.linspace(-a,a,1000)
plt.xlim(-a,a)
plt.plot (x, myfunc(x), x, interpolate.splev(x, tck), pts, vals, "x")
```

the spline is represented as a B-spline:  knots t, coefficients s, order k
