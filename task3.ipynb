```python
import numpy as np
from numpy import cos
import matplotlib.pyplot as plt
import math
```

```python
def t(n,x):
    if n == 0:
        return np.ones(len(x))
    elif n == 1:
        return x
    else:
        return 2*x*t(n-1,x) - t(n-2,x)

xvalues = np.linspace(-2,2,100)

for i in range(0,5):
    plt.plot(xvalues,t(i,xvalues),label=str(i)+". polynomial")

plt.yscale('symlog')
plt.legend()
plt.show()
```
